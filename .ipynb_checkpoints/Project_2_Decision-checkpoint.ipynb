```python
#best K does not consume negative numbers 

encoded_df = raw_data.copy()

encoded_df['pdays'].replace(to_replace=-1,value=np.nan, inplace=True)
encoded_df.loc[encoded_df.balance <= 0, 'balance'] = 0
```

```python
#encode each categorial series
for column in encoded_df:

    if encoded_df[column].dtype == 'O':
    
        s_to_enc = encoded_df[column]
        mapping = {k: v for v, k in enumerate(s_to_enc.unique())}
        encoded_df[column] = s_to_enc.map(mapping)
        
    else: 
        pass
```

```python
df = encoded_df.dropna()
from operator import itemgetter
from collections import OrderedDict
import numpy as np
np.set_printoptions(suppress=True)

x_col_names = list(df.drop('y', axis=1).columns.values)
x_col_key = range(len(x_col_names))

hash_kbest =  dict(zip(x_col_key, x_col_names))

from sklearn.feature_selection import SelectKBest, chi2
from sklearn.feature_extraction import text

X, y = df.drop('y', axis=1), df.y
X.shape

X_new = SelectKBest(chi2, k='all')
X_new.fit_transform(X, y)

list_fields = X_new.get_support(indices=True)
list_scores = X_new.pvalues_
#list_scores = X_new.scores_

results = OrderedDict(sorted(zip(list_fields,list_scores), key=itemgetter(1), reverse=True))
```

```python
print 'Field', '    |       Chi2'
print '-----------------------------'
    
for i in range(1, len(results)-1):

    print str(hash_kbest[results.items()[i+1][0]]),'  |  ', results.items()[i+1][1]
```
